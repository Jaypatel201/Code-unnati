```python
# x axis values
x = [1,2,3]

#corresponding y axis values 
y = [2,4,1]

#name x
plt.xlabel('sales')

#Name y
plt.ylabel('profit')

# giving title
plt.title('my first graph using matplotlib')

plt.plot(x,y)
```

# Sub ploting

```python
#plot 1

x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

#plot 2
z = np.array([0,1,2,3])
w = np.array([10,20,30,40])

plt.subplot(2,1,1)
plt.plot(x,y)

plt.subplot(2,1,2)
plt.plot(z,w)
```

```python
plt.subplot(2,2,4)
plt.plot(x,y)

plt.subplot(2,2,3)
plt.plot(z,w)
```

```python
# Importing mplot3D toolkit , numpy and matplotlib

from mpl_toolkits import mplot3d 
import numpy as np
import matplotlib.pyplot as plt 

fig = plt.figure()

# syntax for 3D projection 
ax = plt.axes(projection = '3d')

#definding all 3 axes 
z = np.linspace(0,1,100)
x = z*np.sin(25*z)
y = z*np.cos(25*z)

#plotting 
ax.plot3D(x,y,z,'r')
ax.set_title('3D line plot')

plt.show()
```
